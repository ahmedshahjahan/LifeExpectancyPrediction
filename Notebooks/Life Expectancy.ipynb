```python
fig, ax = plt.subplots(figsize=(12,8))
sns.lineplot(x=df_BD.Year, y=df['Life expectancy'], color='r', label='Life Expectancy')
sns.lineplot(x=df_BD.Year, y=df['Adult Mortality'], color='b', label='Adult Mortality')
sns.lineplot(x=df_BD.Year, y=df['infant deaths'], color='g', label='Infant Death')
plt.title('Life Expectancy of Bangladesh from 2000 to 2015')
plt.legend()
plt.savefig('Life_Expectancy_VS_Mortality.png', dpi=300)
plt.show()
```
